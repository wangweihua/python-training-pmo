# 小结

介绍了列表、元组、集合、字典：

| 结构      | 同质 | 可变 | 重复 |  有序  | 映射 |
|------------:|-----:|-----:|-----:|-----:|-----:|
|列表（list） | 否   |  是 | 是   | 是 |  否  |
|元组（tuple）| 否   |  否 | 是   | 是 |  否  |
|集合（set）  | 否   |  是 | 否   | 否 |  否  |
|字典（dict） | 否   |  是 | 是   | 否 |  是  |
|字符串（str）| 是   |  否 | 是   | 是 |  否  |

```python
import itertools

a1 = ['同质', '异质']
a2 = ['可变', '不可变']
a3 = ['独一', '重复']      
a4 = ['有序', '无序']
a5 = ['映射', '无映射']
for i, xa in enumerate(itertools.product(a1, a2, a3, a4, a5)):
    print('结构{}：{}'.format(i, xa))
```
